```python
# Assuming your DataFrame is named df and it contains the necessary columns

# Step 1: Filter data for the year 2022
crime_2022 = df[df['year'] == 2022]

# Step 2: Count the number of crimes in each police district for 2022
district_crime_counts = crime_2022['district'].value_counts()

# Step 3: Identify the district with the highest and lowest number of crimes
district_most_crimes = district_crime_counts.idxmax()
district_least_crimes = district_crime_counts.idxmin()

most_crimes_count = district_crime_counts.max()
least_crimes_count = district_crime_counts.min()

print(f"District with the most crimes in 2022: {district_most_crimes} with {most_crimes_count} crimes")
print(f"District with the least crimes in 2022: {district_least_crimes} with {least_crimes_count} crimes")

# Step 4: Plot the crime distribution across different districts
district_crime_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Number of Crimes by Police District in 2022')
plt.xlabel('Police District')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Assuming 'df' is your DataFrame containing Chicago crime data

# Extract year from the 'date' column
df['year'] = pd.to_datetime(df['date']).dt.year

# Group data by year and count total crimes each year
total_crimes_per_year = df.groupby('year').size()

# Plot the trend of total crimes over the years
plt.figure(figsize=(10, 6))
plt.plot(total_crimes_per_year.index, total_crimes_per_year.values, marker='o', linestyle='-')
plt.title('Total Crimes Across the Years')
plt.xlabel('Year')
plt.ylabel('Total Number of Crimes')
plt.grid(True)
plt.show()

# Group data by both year and primary type to analyze individual crimes
crimes_by_type_per_year = df.groupby(['year', 'primary_type']).size().unstack()

# Plot the trend of specific crimes over the years (you can choose specific crimes of interest)
plt.figure(figsize=(12, 8))
crimes_of_interest = ['THEFT', 'BATTERY', 'NARCOTICS']  # Example crimes
for crime_type in crimes_of_interest:
    plt.plot(crimes_by_type_per_year.index, crimes_by_type_per_year[crime_type], marker='o', label=crime_type)

plt.title('Trend of Specific Crimes Across the Years')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Assuming your DataFrame is named df and it contains the necessary columns

# Step 1: Extract month from the date column
df['month'] = pd.to_datetime(df['date']).dt.month

# Step 2: Group data by month and count the number of crimes
crime_counts_by_month = df.groupby('month').size()

# Step 3: Determine months with the highest and lowest crime counts
month_most_crime = crime_counts_by_month.idxmax()
month_least_crime = crime_counts_by_month.idxmin()

most_crime_count = crime_counts_by_month.max()
least_crime_count = crime_counts_by_month.min()

print(f"Month with the most crime: {month_most_crime} with {most_crime_count} crimes")
print(f"Month with the least crime: {month_least_crime} with {least_crime_count} crimes")

# Step 4: Plot the variation in crime counts across months
crime_counts_by_month.plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Number of Crimes by Month')
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```
